```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mat
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
df = df.head(10)
df
```

```python
#barplot for Evaporation and Sunshine
sb.barplot(x = 'Evaporation' , y = 'Sunshine', data = df)
mat.title('Bar Plot for Evaporation and Sunshine')
sb.despine()
```

```python
#boxplot for Evaporation and Sunshine
sb.boxplot(x = 'Evaporation' , y = 'Sunshine', data = df)
mat.title('Box Plot for Evaporation and Sunshine')
sb.despine()
```

```python
#point plot for minimum temperature and maximum temperature
sb.pointplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mat.title('point Plot for minimum temperature and maximum temperature')
sb.despine()
```

```python
#Strip plot for Evaporation and Sunshine
sb.stripplot(x = 'Evaporation', y = 'Sunshine', data = df)
mat.title('Strip plot for Evaporation and Sunshine')
sb.despine()
```

```python
#Violin plot for Min Temperature and Max Temperature
sb.violinplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mat.title('Violin Plot for Minimum Temperature and Maximum Temperature')
sb.despine()
```

```python
#Joint plot for Min Temperature and Max Temperature
sb.jointplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mat.title('Joint Plot for Minimum Temperature and Maximum Temperature')
sb.despine()
```

```python
#Swarm plot plot for Min Temperature and Max Temperature
sb.swarmplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mat.title('Swarm Plot for Minimum Temperature and Maximum Temperature')
sb.despine()
```

```python
sb.lmplot(x = 'Humidity9am' , y = 'Humidity3pm', data = df)
mat.title('LM Plot for Humidity_9am and Humidity_3pm')
sb.despine()
```

```python
sb.lmplot(x = 'Humidity9am' , y = 'Humidity3pm', hue = 'Sunshine', data = df)
mat.title('LM Plot for Humidity_9am and Humidity_3pm with Sunshine')
sb.despine()
```

```python
#Pie Chart
mat.pie(x = 'Humidity9am', data = df)
mat.title('Pie Chart representation for Humidity_9am')
sb.despine()
```

```python
sb.distplot(df['Humidity9am'] , kde = False)
mat.title('distplot for Humidity_9am')
sb.despine()
```

```python
sb.pairplot(df , kind = 'reg')
mat.title('Pair Plot')
sb.despine()
```

```python
sb.factorplot(x = 'Humidity9am' , y = 'Humidity3pm' , hue = 'Sunshine', data = df)
mat.title('lmplot for Humidity9am vs Humidity3pm with hue')
sb.despine()
```

```python
mat.pie(x = 'Sunshine', data = df)
mat.title('Pie Chart for Sunshine')
sb.despine()
```

```python
sb.boxplot(x = 'MaxTemp' , y = 'MinTemp', hue = 'RISK_MM', data = df)
mat.title('Box Plot for MaxTemp vs MinTemp with hue')
sb.despine()
```
