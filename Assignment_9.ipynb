```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
%matplotlib inline
```

Question 1: Concatenate the datasets. 

```python
#titanic train datasets
df  = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
#titanic test datasets
df1 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
```

```python
#concat datasets
concat = [df, df1]
report = pd.concat(concat)
report.head()
```

Question 2: Remove NaN values from the column ‘Survived’. 

```python
#find all null values
report.isnull().sum()
```

```python
#drop Null values from survived feature
remove_null = report.dropna(subset=['Survived'])
remove_null.head()
```

```python
#find null value in dataset after drop
remove_null.isnull().sum()
```

Question 3: Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not.

```python
#new column 'Has_Cabin' that tells you whether passengers had a cabin or not
report['Has_Cabin']=np.where(report['Cabin'].isnull(),'No','Yes')
report.head()
```

Question 4: Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc.

```python
#extract Title from the data set
report['Title'] = report.Name.str.extract(r',\s*([^\.]*)\s*\.',expand =False)
report.head()
```

Question 5: Impute the missing values using .fillna() method

```python
#fill all missing values
n=report.fillna(report.mean())
n=report.fillna(report.mode)
n.isnull().sum()
```

Question 6: Create two new columns “Categorical_age” and “Categorical_fare”, where you will put people that are within a certain range of age or fare in the same bin.

```python
#insert two columns Categorical_age and Categorical_fare
report['Categorical_age'] = pd.qcut(report['Age'],q=4,labels = False)
report['Categorical_Fare'] = pd.qcut(report['Age'],q=4,labels = False)
report.head()
```

Question 7: Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.

```python
#drom some features of datasets
new_dataset=report.drop(['Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'],axis =1)
new_dataset.head()
```

Question 8: Transform all variables into numeric ones.

```python
#get_dummies to convert into numerical values
d=pd.get_dummies(report)
f=d.fillna(d.min())
#now report is f.
```

Questin 9:Now we need training set and testing set. Use train_test_split() to split your dataset. 
X(Features) will be the dataset values without the column “Survived” and y(Targets) will be your “Survived” column values.

```python
y = f['Survived']
X = f.drop(['Survived'],axis=1)
```

```python
regressor = LogisticRegression()
```

```python
#perform split on training data
x_train,x_test,y_train,y_test  = train_test_split(X,y,test_size = 0.3, random_state = 2)
```

Question 10:Use logistic regression by passing your X_train and y_train.

```python
#logistic regression
regressor.fit(x_train,y_train)
```

```python
#get predicted data
pred = regressor.predict(x_test)
```

```python
#obtain data frame from predicted value and y_test
result = pd.DataFrame({"Predicted":pred,"Actual":y_test})
result.head()
```

Question 11:Print the accuracy score.

```python
#accuracy score
regressor.score(x_test,y_test)*100
```
